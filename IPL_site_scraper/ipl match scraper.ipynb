```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
```

```python
# call the webdriver and start the crome
driver = webdriver.Chrome(executable_path= r"C:\api\chromedriver.exe")
```

```python
# define one base_url of the sites
base_url = 'https://www.iplt20.com/matches/results/'
#list of all the sessions
session_list = list(range(2008,2024))
# create one dictionary of all the sessions
all_session_url = {}
for i in range(2008,2024):
    all_session_url[i] = base_url+str(i)
```

```python
# creating one fuction which will take soup of one line and return all the details of the single match
def get_match_detail(li_soup, url):
    match_number = li_soup.find("span", class_ = "vn-matchOrder ng-binding ng-scope").text.replace("Match ","")
    venue = li_soup.find("p", class_ = "ng-binding").text
    date_time = li_soup.find("div", class_ = "vn-matchDateTime ng-binding").text.strip()
    winning_team_title = li_soup.find("div", class_ = "vn-ticketTitle ng-binding ng-scope").text
    team_stats = li_soup.find_all("div", class_="vn-teamTitle")
    first_team_stats = team_stats[0].text.strip()
    second_team_stats = team_stats[1].text.strip()
    link = li_soup.find("a", class_="vn-matchBtn ng-scope", href = True)["href"]
    print(match_number,'successfully scraped')
    return {"session": url[-4:], "matchnumber": match_number, "venue": venue, "dateandtime": date_time, "winningteam": winning_team_title,
           "firstteam": first_team_stats, "secondtime": second_team_stats, "link": link}
```

```python
# defining one function which will take one url and return all the details of the one session
def get_season_details(url):
    # get the url using driver
    driver.get(url)
    sleep(10)
    # get all the html data from  the site
    html = driver.page_source
    # convert all the html data into a soup
    soup = BeautifulSoup(html)
    # finding all the line tag conatin all the info about the match
    list_of_li = soup.find("ul",{"id":"team_archive"}).find_all("li", class_ = "ng-scope")
    # creating one empty list (to store all the data of the matches)
    session_list = []
    # using one for loop to get line tag one by one from the list_of_li and call the get_match_detail func to scrap the detail
    # of the match
    for i in list_of_li:
        session_list.append(get_match_detail(i, url))
    print("*****************************")
    print(url[-4:], "scrapped successfully")
    print("*****************************")
    return session_list # return whole session info 
    
    
```

```python
all_session_data = []
for i in all_session_url.values():
    all_session_data.append(get_season_details(i))
```

```python
df = pd.DataFrame(all_session_data[0])
```

```python
for i in all_session_data[1:]:
    new_df = pd.DataFrame(i)
    df = pd.concat([df, new_df], ignore_index=True)
```

```python
df.to_csv('matches.csv')
```
