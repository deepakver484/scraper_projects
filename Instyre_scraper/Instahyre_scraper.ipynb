```python
# importing libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
# opeing chrome browser  with the help of webdriver
driver = webdriver.Chrome(executable_path= r"C:\api\chromedriver.exe")
driver.maximize_window() #maximising the window
```

```python
#here you can change python with the multiple skills to get the jobs related to that skill only
driver.get('https://www.instahyre.com/python-jobs')
```

```python
'''Here I am making one fucntion whihc will take soup of company as a input and return all the info of 
that company in the form of dictonary'''
def find_info(soup_of_company):
    try:
        company_name =  soup_of_company.find('div', class_="employer-job-name").text.strip()
    except:
        company_name = "not given"
    try:
        loc = soup_of_company.find('div', class_="employer-locations").text.strip()
    except:
        loc = "not given"
    try:
        founded, employees = soup_of_company.find('div', class_="employer-info").text.strip().split('\n\nâ€¢\n\n')
    except:
        founded, employees = "not given", "not given"
    try:
        about = soup_of_company.find('div', class_="employer-notes ng-binding ng-scope").text.strip()
    except:
        about = "not given"
    try:
        skills = ",".join([i. text for i in soup_of_company.find('div', class_="job-skills ng-scope").find_all("li")])
    except:
        skills = "not given"
    try:
        link = soup_of_company.find('a', href = True)['href']
    except:
        link = "not given"
    return {"name":company_name, 'location': loc, "founded": founded, "employees": employees, "about":about, "skills": skills, 'link': link }
```

```python
# here is simple code to get the soup of all company for 50 pages
job_list = []
for i in range (50):
    if i>=1:
        driver.find_element(By.XPATH, "/html/body/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/div[11]/li[12]").click()
        sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html)
    list_comp = soup.find_all("div", class_ = "employer-row")
    job_list += list_comp
    print(f"page {i} extracted sucessfully")
    
```

```python
# creating pandas dataframe with the soup we are getting 
df = pd.DataFrame(job_list, columns = ['sep1', 'main', 'sep2'])
```

```python
# extracting info of the dictonary and creating multiple columns out of it
df[['name', 'location', 'founded', 'employees', 'about', 'skills', 'link']] = df['main'].apply(lambda x: pd.Series(find_info(x)))
```

```python
# droping useless columns from the dataframe
df.drop(columns = ['sep1', 'main', 'sep2'], inplace = True)
```
