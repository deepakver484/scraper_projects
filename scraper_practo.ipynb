```python
# Imporing libraries and modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# creating two lists for cities and specility 
cities = ['Mumbai', 'Pune', 'Delhi', 'Ahmedabad', 'Surat', 'Chennai', 'Kolkata', 'Bangalore', 'Hyderabad']
speciality=["Dentist","Gynecologist","Pediatrician","Orthopedist","Dermatologist","Physiotherapist","Infertility Specialist",
            "Ophthalmologist","Urologist","Psychiatrist","Cardiologist","Dietitian","Gastroenterologist","Neurologist",
            "Neurosurgeon","Pulmonologist","Bariatric Surgeon","Rheumatologist","Chiropractor"]
```

```python
# passing one link and creat links for all the cities and the speciality 
links = [ f'https://www.practo.com/search/doctors?results_type=doctor&q=%5B%7B%22word%22%3A%22{j}%22%2C%22autocompleted%22%3Atrue%2C%22category%22%3A%22subspeciality%22%7D%5D&city={i}&page='
for i in cities for j in speciality]
```

```python
''' Created one function which will take link as an input and return list of html of doctors info present for that city and 
speciality'''
def links_scraper(link):
    list_of_doctors_info = []
    i = 1
    lis = [1]
    while len(lis) !=0:
        link_f = f"{link}{i}"
        soup = BeautifulSoup(requests.get(link_f).text)
        lis = soup.find_all('div', class_='listing-doctor-card')
        list_of_doctors_info += lis
        i += 1
    return list_of_doctors_info
```

```python
''' Created one function which will take links as input and call links_scraper function to scrap all the links data and 
return one list info containing html of all doctors info'''
def info_links(links):
    info = []
    for i in range(len(links)):
        info += links_scraper(links[i])
        print(i, 'extracted')
    return info
    
```

```python
# calling info_links function to get all the html data for all links
info = info_links(links)
```

```python
''' this function will take html code for one doctor and extract all the usefull information from that html code and reurn a dictonary
of containing that info '''
def doctor(html_soup):
    try:
        name = html_soup.find('h2', class_ = 'doctor-name').text
    except:
        name = None
    try:
        speacility = html_soup.find('div', class_ = 'info-section').find('div', class_='u-d-flex').text
    except:
        speacility = None
    try:
        experience = html_soup.find('div', attrs={'data-qa-id': 'doctor_experience'}).text.split('\xa0')[0]
    except:
        experience = None
    try:
        locality, city = html_soup.find('div', class_='u-bold u-d-inlineblock u-valign--middle').text.split(',')
    except:
        locality, city = None, None
    try:
        institute = html_soup.find('span', class_='u-c-pointer u-t-hover-underline').text
    except:
        institute = None
    try:
        fee = html_soup.find('span', attrs={'data-qa-id': 'consultation_fee'}).text
    except:
        fee = None
    try:
        recommendation = html_soup.find('span', attrs={'data-qa-id': 'doctor_recommendation'}).text
    except:
        recommendation = None
    try:
        feedback = html_soup.find('span', attrs={'data-qa-id': 'total_feedback'}).text.split('\xa0')[0]
    except:
        feedback = None
    try:
        booking_fee = html_soup.find('button', class_='u-t-capitalize u-bold u-round-corner--large c-btn--dark-medium').text
    except:
        booking_fee= None
    try:
        video_booking = html_soup.find('button', class_='u-t-capitalize u-bold u-round-corner--large c-btn--purple-light u-border--thick')
    except:
        video_booking = None
    return {'name': name, 'speacility': speacility, 'experience': experience, 'locality' : locality, 'city': city, 
            'institute': institute, 'fee': fee, 'recommendation' : recommendation, 'feedback': feedback,
            'booking_fee' : booking_fee, 'video_booking': video_booking}
```
