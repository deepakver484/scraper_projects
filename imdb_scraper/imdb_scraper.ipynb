```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link = 'https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012'
base_link = 'https://www.imdb.com'
```

```python
desc_link = 'https://www.imdb.com/search/title/?title_type=feature&year=1950-01-01,2012-12-31&sort=num_votes,desc'
links = []

for i in range(1, 4001):
    new_link = f'https://www.imdb.com/search/title/?title_type=feature&year=1950-01-01,2012-12-31&sort=num_votes,desc&start={50*i+1}'
    links.append(new_link)
links
```

```python
def page_link(link):
    soup = BeautifulSoup(requests.get(link).text)
    movies_soup = soup.find_all('div', class_ = 'lister-item-content')
    return movies_soup
```

```python
def movie(movies_soup):
    try:
        movie_link = base_link + movies_soup.find('h3', class_= 'lister-item-header').find('a', href= True)['href']
    except:
        movie_link = None
    try:
        movie_name = movies_soup.find('h3', class_= 'lister-item-header').find('a').text
    except:
        movie_name = None
    try:
        year = movies_soup.find('h3', class_= 'lister-item-header').find('span', class_ = 'lister-item-year text-muted unbold').text[1:-1]
    except:
        year = None
    try:
        run_time = movies_soup.find('span', class_ = 'runtime').text
    except:
        run_time = None
    try:
        genre = movies_soup.find('span', class_ ='genre').text.strip()
    except:
        genre = None
    try:
        imdb_rating = movies_soup.find('strong').text
    except:
        imdb_rating = None
    try:
        metascore = movies_soup.find('span', class_ = 'metascore favorable').text.strip()
    except:
        metascore = None
    try:
        cast =  movies_soup.find('p', class_ = "").text.strip().replace('\n', '')
    except:
        cast = None
    try:
        director = cast[cast.index(':')+1:cast.index('|')]
    except:
        director = None
    try:
        actors = cast[cast.index('Stars:')+6:]
    except:
        actors = None
    try:
        votes_gross = movies_soup.find('p', class_ ='sort-num_votes-visible').text.replace('\n', '')
    except:
        votes_gross = None
    try:
        votes = votes_gross[votes_gross.index(':')+1:votes_gross.index('|')]
    except:
        votes = None
    try:
        gross = votes_gross[votes_gross.index('Gross:')+6:]
    except:
        gross = None
    
    return {'link': movie_link, 'name':movie_name, 'year': year, 'runtime': run_time, 'genre': genre, 'rating': imdb_rating, 
           'metascore': metascore, 'director' : director, 'actors': actors, 'votes': votes, 'gross': gross}
    
```

```python
movie_list = []
for i in links:
    movies_soup = page_link(i)
    for m in range(50):
        movie_soup = movies_soup[m]
        try:
            movie_list.append(movie(movie_soup))
        except:
            continue
    print(f'{len(movie_list)}: movies extracted')
    
```

```python
df = pd.DataFrame(movie_list)
```
