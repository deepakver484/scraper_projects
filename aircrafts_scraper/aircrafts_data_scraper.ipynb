```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
links = BeautifulSoup(requests.get("https://www.airlines-inform.com/commercial-aircraft/").text).find('section', class_ ="ad-infoblok").find_all('a', href = True)
```

```python
final_links = [i['href'] for i in links]
final_links
```

```python
table = BeautifulSoup(requests.get(final_links[0]).text).find('table', class_='aircraft')
```

```python
# Create a Pandas DataFrame to store the table data
all_data = {}
def table_data(link):
    try:
        table = BeautifulSoup(requests.get(link).text).find('table', class_='aircraft')
        data = []
        for row in table.find_all('tr'):
            row_data = []
            for cell in row.find_all('td'):
                row_data.append(cell.get_text())
            data.append(row_data)
        all_data[link] = data
        print('done')
    except:
        print('not')
    return 'done'
```

```python
df = pd.DataFrame({'links': final_links})
df['links'].apply(table_data)
```
