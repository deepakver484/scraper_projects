```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
def books_link(pagelink):
    soup = BeautifulSoup(requests.get(pagelink).text)
    link = soup.find_all('li',class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
    baseURL = 'https://books.toscrape.com/catalogue/'
    book_link = []
    for i in link:
        lnk = i.find('a',href = True)['href']
        book_link.append(f'{baseURL}{lnk}')
    return book_link
```

```python
def book_info(book_link):
    book_soup = BeautifulSoup(requests.get(book_link, timeout=10).text)
    UID = book_soup.find('table',class_ = 'table table-striped').find_all('td')[0].text
    Book_name = book_soup.find('ul',class_ = 'breadcrumb').find_all('li')[3].text.strip()
    genre = book_soup.find('ul',class_ = 'breadcrumb').find_all('li')[2].text.strip()
    price = book_soup.find('p',class_ = 'price_color').text.replace('Â£','')
    Number_of_reviews = book_soup.find('table',class_ = 'table table-striped').find_all('td')[-1].text
    ratings = str(book_soup.find('p',class_ = 'star-rating')).split()[2].replace('">','')
    return {'UID':UID, "Book_name": Book_name, "genre": genre, 'price':price, 'Number_of_reviews': Number_of_reviews,
            'ratings': ratings}
```

```python
def books_info(books_link):
    info = []
    for i in books_link:
        try:
            info.append(book_info(i))
            print(f' {books_link.index(i)} links extracted')
        except:
            print(f' {books_link.index(i)} links failed')
            continue
    return info
```

```python
def pages(Number_of_pages):
    book_link = []
    for i in range(1,Number_of_pages+1):
        try:
            pagelink = f'https://books.toscrape.com/catalogue/page-{i}.html'
            book_link.append(books_link(pagelink))
            print(f'page {i} links extracted')
        except:
            continue
    return book_link    
```

```python
li = pages(50)
```

```python
flat_list = [item for sublist in li for item in sublist]
```
